```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data ={'hours':[1,2,3,4],
        'marks':[20,40,50,60]
}
```

```python
df=pd.DataFrame(data)
```

```python
print(df)
```

```python
X=df['hours'].values
y=df['marks'].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X.reshape(-1, 1), y)
```

```python
slope =model.coef_[0]
```

```python
intercept = model.intercept_
```

```python
print(f"slope(m):{slope}")
print(f"intercept(b):{intercept}")
print(f"equation of the regression line: y ={slope:.2f}x +{intercept:.2f}")
```

```python
df['predicted_marks']= model.predict(X.reshape(-1, 1))
```

```python
df.head()
```

```python
plt.scatter(df['hours'],df['marks'],color='blue',label='actual_marks')
```

```python
plt.plot(df['hours'],df['predicted_marks'],color='red',label='Regression line')
plt.xlabel('Hours Studied')
plt.ylabel('marks')
plt.title('linear regression')
plt.legend()
plt.show()
```
